# Q2. Max, min , mean of asked amount,asked equity, asked valuation?

# Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise? 

```python
episodewise_max = data.groupby('episode_number')[["pitcher_ask_amount", "ask_equity", "ask_valuation"]].max()
episodewise_max
```

```python
episodewise_min = data.groupby('episode_number')[["pitcher_ask_amount", "ask_equity", "ask_valuation"]].min()
episodewise_min
```
